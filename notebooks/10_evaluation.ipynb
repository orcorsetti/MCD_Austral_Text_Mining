# 10 — Evaluación del Sistema

Evaluación cuantitativa del pipeline RAG completo.

**Métricas de retrieval:**
- Precision@K (K=5, 10)
- Recall@K
- MRR (Mean Reciprocal Rank)
- nDCG

**Métricas RAG (RAGAS):**
- Context Precision / Recall
- Faithfulness
- Answer Relevance

**Ground truth:** 10-20 perfiles ficticios con estudios aplicables marcados manualmente  
**Comparación baseline:** keyword matching vs retrieval semántico

**Input:** Resultados de nb08 y nb09  
**Output:** Tabla comparativa de métricas y análisis de errores
