# 09 — Generación con LLM

Generación de explicaciones de elegibilidad para los estudios candidatos.

**Objetivos:**
- Diseño del prompt con perfil del paciente + criterios de los Top-N estudios
- Generación de justificación por estudio:
  - Criterios que matchean
  - Criterios no evaluables por falta de datos en el formulario
  - Posibles incompatibilidades a verificar
- Comparar modelos:
  - API Anthropic (Claude)
  - `mistralai/Mistral-7B-Instruct-v0.3` (local)
  - `meta-llama/Meta-Llama-3-8B-Instruct` (local)

**Input:** Top-N estudios con scores (nb08)  
**Output:** JSON con lista de estudios + explicaciones
